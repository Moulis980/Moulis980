```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_excel("/content/BMW.xlsx",header=None,skiprows=1)
df.drop(0,axis="columns",inplace=True)
df.columns=["Mileage","Age","Selling_Price"]
print(df.to_string())
```

```python
#plt.plot(df["Mileage"],df["Age"],color="r")
plt.figure(figsize=(10,5))
plt.plot(df["Mileage"],df["Selling_Price"],color="b",marker="o")
plt.xlabel("Mileage")
plt.ylabel("Selling Price")
plt.show()
```

```python
x=df[["Mileage","Age"]]
y=df.Selling_Price
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3)
```

```python
gr=LinearRegression()
gr.fit(X_train,Y_train)
```

```python
X_test
```

```python
gr.predict(X_test)
```

```python
Y_test
```

```python
gr.score(X_test,Y_test)
```
